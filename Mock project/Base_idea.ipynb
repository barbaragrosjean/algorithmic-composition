```python
#fonction utiles 

def selectRandom(names):
    return random.choice(names)

def play(score) : 
    midi.realtime.StreamPlayer(score).play()
    
#liste des choix dans les bonnes notes des accords

CHORDS1 =['C4','E4', 'G4','B4','C5','E5', 'G5','B5' ]

#sequence melodique initialiser comme sequence vide

melody_pitch0 = []
melody_pitch1 = []
melody_pitch2 = []
melody_pitch3 = []
melody_pitch4 = []
```

```python
#generation de la sequence 0

melody0 = stream.Part()

#par prudance
melody0.clear()
melody_pitch0.clear()

for n in range(4) :
    melody_pitch0.append(selectRandom(CHORDS1))

for n in range (0,10) : 
        for i in range(len(melody_pitch0)):
            melody0.append(note.Note(pitch = melody_pitch0[i], quarterLength = 2))
            
melody0.show()
#play(melody0)
```

```python
#generation de la sequence 1 
melody1 = stream.Part()

melody1.clear()
melody_pitch1.clear()

#initialisation avec les premieres notes de melody_pitch0
melody_pitch1 = melody_pitch0

for n in range(8) :
    melody_pitch1.append(selectRandom(CHORDS1))


for n in range (0,10) :
    if n < 1 :
        for k in range(0,8) :
            melody1.append(note.Rest(quaterLenght =8))
    else :  
        for i in range(len(melody_pitch1)):
            melody1.append(note.Note(pitch = melody_pitch1[i], quarterLength = 1))
            
melody1.show()
#play(melody1)
```

```python
#generation de la sequence 2 
melody2 = stream.Part()

melody2.clear()
melody_pitch2.clear()

melody_pitch2 = melody_pitch1

for n in range(16) :
    melody_pitch2.append(selectRandom(CHORDS1))


for n in range (0,10) :
    if n < 2 :
        for k in range(0,8) :
            melody2.append(note.Rest(quaterLenght =8))
    else :  
        for i in range(len(melody_pitch2)):
            melody2.append(note.Note(pitch = melody_pitch2[i], quarterLength = 1/2))
            
melody2.show()
#play(melody2)
```

```python
#generation de la sequence 3 
melody3 = stream.Part()

melody3.clear()
melody_pitch3.clear()

melody_pitch3 = melody_pitch2

for n in range(32) :
    melody_pitch3.append(selectRandom(CHORDS1))

#melody3.append(key.KeySignature(-3))

for n in range (0,10) :
    if n < 3 :
        for k in range(0,8) :
            melody3.append(note.Rest(quaterLenght =8))
    else :  
        for i in range(len(melody_pitch3)):
            melody3.append(note.Note(pitch = melody_pitch3[i], quarterLength = 1/4))
    
melody3.show()
#play(melody3)
```

```python
#generation de la sequence 4 
melody4 = stream.Part()

melody_pitch4.clear()
melody4.clear()

melody_pitch4 = melody_pitch3

for n in range(64) :
    melody_pitch4.append(selectRandom(CHORDS1))

for n in range (0,10) :
    if n < 4 :
        for k in range(0,8) :
            melody4.append(note.Rest(quaterLenght =8))
    else :  
        for i in range(len(melody_pitch4)):
            melody4.append(note.Note(pitch = melody_pitch4[i], quarterLength = 1/8))
    
melody4.show()
#play(melody4)
```
